```python
# Librairies
import sys
import pandas as pd
from pathlib import Path

# Racine du projet : remonte jusqu'au dossier qui contient "src"
project_root = Path().resolve()

while not (project_root / "src").exists():
    if project_root == project_root.parent:
        raise FileNotFoundError("Impossible de trouver le dossier 'src'. Vérifie le dossier courant du notebook.")
    project_root = project_root.parent

# Ajouter la racine du projet au Python path
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

# Imports projet
from src.data.data_management import enrich_historical_components, convert_prices_to_eur, MSCIWorldComponentsByDate
from src.data.utils import ParquetSaver

print(f"Project root détecté : {project_root}")
```

```python
# Création de la colonne "exchange_rate" dans le dataframe des informations historiques des composants du MSCI World

input_path = project_root / "data" / "informations_historical_components_msci_world.parquet"
df = pd.read_parquet(input_path)

df = enrich_historical_components(df)
print(df.head(10))

saver = ParquetSaver(subfolder=Path("storage"))
saver.save_parquet(df, "informations_historical_components_msci_world.parquet")
```

```python
# conversion des prix historiques des composants du MSCI World en EUR

prices_df = pd.read_parquet(project_root / "data" / "historical_prices_components_msci_world.parquet")
currencies_df = pd.read_parquet(project_root / "data" / "historical_currencies_eur_components_msci_world.parquet")
info_df = pd.read_parquet(project_root / "data" / "storage" / "informations_historical_components_msci_world.parquet")

prices_df_eur = convert_prices_to_eur(prices_df, info_df, currencies_df)

saver = ParquetSaver(subfolder=Path("storage"))
saver.save_parquet(prices_df_eur, "historical_prices_eur_components_msci_world.parquet")
```

```python
# transfere des derniers fichiers dans le dossier "storage" qui sont corrects
weights_df = pd.read_parquet(project_root / "data" / "historical_weights_components_msci_world_by_month.parquet")
saver = ParquetSaver(subfolder=Path("storage"))
saver.save_parquet(weights_df, "historical_weights_components_msci_world_by_month.parquet")
```
